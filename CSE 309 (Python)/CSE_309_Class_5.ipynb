**LOOP**

print: 0-9
JAVA code:
```
for(int i=0; i<10; i++) {
  System.out.println(i)
}

int i=0;
while (i<10) {
  System.out.println(i);
  i++;
}
```



````
i = 0
while (i<10) :
  print(i)
  i = i + 1
````

````
i=1
sum = 0
while i<11 :
  if(i == 10):
    print(i, end=' ')
  else:
    print(i, "+" , end=' ')
  sum = sum + i
  i += 1
print("=", sum)
````

````
start = int(input("Start: "))
end_p = int(input("End: "))

sum = 0
i=start

while i<end_p+1 :
  if(i == end_p):
    print(i, end=' ')
  else:
    print(i, "+" , end=' ')
  sum = sum + i
  i += 1
print("=", sum)
````
